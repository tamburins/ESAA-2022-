### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
users=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",  delimiter='|')
users
```

### Step 3. Assign it to a variable called users.

```python
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

(Define 'function' to solve this probelm)

```python
genders = users.groupby("occupation")["gender"].value_counts(normalize=True)*100
genders.head()
gendersn = pd.DataFrame(genders)
gendersn.rename(columns={"gender":"percent"}, inplace=True)
gendersn.head()
new_gen = pd.DataFrame(gendersn.reset_index(level=["occupation", "gender"]))
new_gen.head()
males = new_gen[new_gen["gender"] == "M"].sort_values(by="percent",ascending=False)
males
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").age.agg([min, max])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"]).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
def males_ratio(k):
  a = k[k['gender']=='M'].groupby('occupation')['gender'].count() / k.groupby('occupation')['gender'].count()
  a.sort_values(ascending=False, inplace=True)

  return a

males_ratio(users)
```
