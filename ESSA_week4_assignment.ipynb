## 다변량 그래프 그리기

3개 이상의 변수를 사용한 그래프

```python
def recode_sex(sex): #문자열은 산점도의 구분용이 될 수 없어 정수로 치환
  if sex=='Female':
    return 0
  else:
    return 1

tips['sex_color']=tips['sex'].apply(recode_sex) #위의 함수가 반환한 값 별도 변수로 출력

```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
scatter_plot=plt.figure()

%matplotlib inline 
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
x=tips['total_bill'], y=tips['tip'], s=tips['size'] * 10, c=tips['sex_color'], alpha=0.5)
axes1.set_title('totalbill vs tip colored by sex and sized by size')
axes1.set_xlabel('totalbills')
axes1.set_ylabel('tip')

scatter_plot
```

## **04-3 seaborn 라이브러리 자유자재로 사용하기**

```python
#단변량 그래프 그리기 - 히스토그램

import seaborn as sns
tips=sns.load_dataset("tips") #tips 호출

ax=plt.subplots()
ax=sns.displot(tips['total_bill'])
ax.set_title('totalbill histogram with density plot') #기본틀과 히스토그램
```

```python
ax=plt.subplots()
ax=sns.displot(tips['total_bill'], kde=False) #kde로 밀집도 그래프 온오프
ax.set_title('total bill histogram')
ax.set_xlabel('total bill')
ax.set_ylabel('frequency')
```

```python
#밀집도 그래프만 나타내기
ax=plt.subplots()
ax=sns.displot(tips['total_bill'], hist=False)
ax.set_title('totalbill density')
ax.set_xlabel('totalbill')
ax.set_ylabel('unit probability')
ax
```

```python
#displot에 rug인자를 추가하여 양탄자 그래프도 한꺼번에 그리기
ax=plt.subplots()
ax=sns.displot(tips['total_bill'], rug=True)
ax.set_title('totalbill rug graph')
ax.set_xlabel('totalbill')
ax
```

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips)
```

```python
ax.set_title('totalbill density')
ax.set_xlabel('totalbill')
ax.set_ylabel('unit probability')
```

```python
#다양한 종류의 이변량 그래프
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip', data=tips)
ax.set_title('scatterplot of total bill n tip')
ax.set_xlabel('total billl')
ax.set_ylabel('tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('scatterplot of total bill n tip')
ax.set_xlabel('total billl')
ax.set_ylabel('tip')
```

```python
#산점도와 히스토그램 한번에 그리기 with joinplot
joint= sns.jointplot(x='total_bill', y= 'tip', data=tips)
joint.set_axis_labels(xlabel='Total bill', ylabel='Tip')
joint.fig.suptitle('joint plot of total bill and tip', fontsize=10, y=1.03)
```

```python
#산점도 그래프의 데이터를 구분하기 위한 육각 그래프
hexbin=sns.jointplot(x="total_bill", y="tip",data=tips,kind="hex")
hexbin.set_axis_labels(xalbel='total bill', yalbel='tip')
hexbin.fig.suptitle('hexbin joint plot of total bill and tip', fontsize=10, y=1.03)
```

## **04-5 seaborn 라이브러리로 그래프 스타일 설정하기**

그래프에 스타일 적용하기

```python
fig, ax= plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('whitegrid')#배경에 가로줄
fig, ax= plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
fig=plt.figure()

seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position=idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)
fig.tight_layout()
```
