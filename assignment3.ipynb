```python
import numpy as np
import matplotlib.pyplot as plt

def getData(filename = '/content/drive/MyDrive///Python  - week3/cereal.txt'):
    """
    Reads the cereal data from the given file name.
    Returns a tuple of column labels, names and data, where
    col_labels Labels of the attributes in the data set
    names      List containing the names of the cereals tested
    data       numpy.ndarray containing the numeric values of the 12 attributes for the 77 cereals
    """
    names = []
    data = []
    col_labels = ["calories (number)", "protein (g)", "fat (g)", "sodium (mg)",
                  "dietary fiber (g)", "complex carbohydrates (g)", "sugars (g)",
                  "display shelf (1, 2, or 3, counting from the floor)", "potassium (mg)",
                  "vitamins and minerals (0, 25, or 100, respectively)",
                  "weight (in ounces) of one serving (serving size)", "cups per serving"]

    with open(filename) as f:
        for l in f:
          fields = l.split()
          names.append(fields[0])
          data.append([float(x) for x in fields[3:]])

    data = np.array(data)
    return col_labels, names, data

def plotFatVsCalories(col_labels, data):
    """
    Plots fat content vs calories
    """

    fat = data[:,2]
    calories = data[:,0]
    plt.plot(fat, calories ,'kx')
    plt.xlabel(col_labels[2])
    plt.ylabel(col_labels[0])
    plt.grid()

    # fit a line through the points
    coefficients = np.polyfit(fat, calories, 1)
    poly = np.poly1d(coefficients)
    plt.plot(fat,poly(fat), 'r-')

#Now execute our functions:
column_labels, productNames, data = getData('/content/drive/MyDrive///Python  - week3/cereal.txt')
plotFatVsCalories(column_labels, data)
```

> **4.** When your colleague hears about your results, he is a bit disappointed. However, he already has a new hypothesis: maybe chocolate is acting as a promoter somehow and the amount of chocolate consumed has some effect? Create some plots and hypothesis tests to assess this new hypothesis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, ttest_ind

# Load the data
data = pd.read_csv('/content/drive/MyDrive///Python  - week3/experiment.csv', header=None, names=['height', 'weight', 'amount_chocolate', 'task_one_score', 'task_two_score'])

# Explore the data
# Scatter plot of amount of chocolate consumed vs. task scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x='amount_chocolate', y='task_one_score', data=data, label='Task One Score')
sns.scatterplot(x='amount_chocolate', y='task_two_score', data=data, label='Task Two Score')
plt.xlabel('Amount of Chocolate Consumed (grams)')
plt.ylabel('Task Scores')
plt.title('Relationship between Chocolate Consumption and Task Scores')
plt.legend()
plt.show()

```

```python
# Correlation analysis
correlation_task1, _ = pearsonr(data['amount_chocolate'], data['task_one_score'])
correlation_task2, _ = pearsonr(data['amount_chocolate'], data['task_two_score'])
print(f"Pearson correlation coefficient between amount of chocolate and Task One Score: {correlation_task1}")
print(f"Pearson correlation coefficient between amount of chocolate and Task Two Score: {correlation_task2}")

# Hypothesis testing
# Compare mean task scores between high and low chocolate consumption groups using t-test
threshold = 50  # Threshold to divide participants into high and low chocolate consumption groups
high_chocolate = data[data['amount_chocolate'] >= threshold]
low_chocolate = data[data['amount_chocolate'] < threshold]

t_statistic, p_value = ttest_ind(high_chocolate['task_one_score'], low_chocolate['task_one_score'])
print(f"T-test results for Task One Score: t-statistic={t_statistic}, p-value={p_value}")

t_statistic, p_value = ttest_ind(high_chocolate['task_two_score'], low_chocolate['task_two_score'])
print(f"T-test results for Task Two Score: t-statistic={t_statistic}, p-value={p_value}")
```
