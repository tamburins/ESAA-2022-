```python
#p1
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace((-1)*np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.grid(False)
positions = [np.pi * (-1), np.pi/(-2), 0, np.pi/2, np.pi] 
labels = ["-π", "-π/2", "0", "+π/2", "π"]
plt.xticks(positions, labels) #in "position", label "labels"
plt.yticks([-1, 0, 1], [-1, 0, 1])
ax1=plt.plot(x, y1, color="r")
ax2=plt.plot(x, y2, color="b")
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_visible(False) #오른쪽 테두리 제거
ax.spines['top'].set_visible(False) #위 테두리 제거
```

p1 설명

필요한 라이브러리를 호출한다. 그 후 sin함수와 cos 함수를 그리기 위한 밑바탕을 위해 ax와 fig를 설정한다. 공통의 -파이에서 +파이를 갖는 변수 x를 설정하고, 파란 선 그래프는 cos, 빨간 선 그래프는 sin이므로  x값에 따라 sin, cos을 갖는 y1과 y2를 설정한다. 

위치를 지정하기 위해 positions와 labels를 설정하고, 이를 그래프에 붙이기 xticks를 사용한다. Yticks도 주어진 범위에맞게 붙이고 주어진 색상 빨강과 파랑에 맞게 그래프를 생성하여 ax1과 ax2에 넣는다.

그림에서의 x,y축은 가운데에 위치하므로 ax.spines를 활용하여 각 축의 위치를 지정한다. 그림에서의 가장자리선은 존재하지 않으므로 마찬갖로 visible에 false옵션을 활용하여 없애준다

```python
#p2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace((-1)*np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.grid(False)
positions = [np.pi * (-1), np.pi/(-2), 0, np.pi/2, np.pi] 
labels = ["-π", "-π/2", "0", "+π/2", "π"]
plt.xticks(positions, labels) #in "position", label "labels"
plt.yticks([-1, 0, 1], [-1, 0, 1])
ax1=plt.plot(x, y1, color="r", label = 'sine')
ax2=plt.plot(x, y2, color="b", label='cosine')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_visible(False) #오른쪽 테두리 제거
ax.spines['top'].set_visible(False) #위 테두리 제거

line1, = ax.plot(x, y2, 'b-')
line2, = ax.plot(x, y1, 'r-')
ax.legend([line1, line2], ['cosine', 'sine'])

ax.plot(np.pi*(2/3),0.866025, 'o', color='r')
ax.plot(np.pi*(2/3),-1/2, 'o', color='b')
plt.plot([np.pi*(2/3),np.pi*(2/3)],[0,0.866025], color="red", linestyle="--")
plt.plot([np.pi*(2/3),np.pi*(2/3)],[0,-1/2], color="blue", linestyle="--")


ax.annotate(r'$\frac{2π}{3}$', xy=(np.pi*(2/3),0.866025), xytext=(np.pi*(4/5), 1.2),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate(r'$\frac{2π}{3}$', xy=(np.pi*(2/3),-1/2), xytext=(np.pi*(1/5), -0.8),
            arrowprops=dict(facecolor='black', shrink=0.05), horizontalalignment='left',
            verticalalignment='bottom')
```

```python
#4
#a
s = np.linspace(0, 2 * np.pi, 100)
t = np.linspace(-0.5, 0.5, 100)
s, t = np.meshgrid(s, t)

#b
x = (1 + t/2 * np.cos(s/2))*np.cos(s)
y = (1 + t/2 * np.cos(s/2))*np.sin(s)
z = t/2 * np.sin(s/2)

#c
x = np.ravel(x)
y = np.ravel(y)
z = np.ravel(z)

fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
ax.plot_trisurf(x, y, z,
                cmap='viridis', edgecolor='none')
```
